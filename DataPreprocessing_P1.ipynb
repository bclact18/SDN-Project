```python
df['NewRecovered'] = df['NewRecovered'].astype(float)
df['Population'] = df['Population'].str.replace(',','').astype(float)
df.interpolate()
```

```python

#df = pd.read_csv('WorldCases2023.csv')
newdf = df.dropna(axis=0, how='any', subset=['Country','NewCases'] )
ct = newdf.loc[0:100, 'Country']
newCase = newdf.loc[0:100, 'NewCases']
print(ct)
print(newCase)
```

```python
plt.plot(ct,newCase)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(ct,newCase)
ax.set_xlabel('Country')
ax.set_ylabel('New Cases')
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(ct,newCase)
ax.set_xlabel('País')
ax.set_ylabel('Casos Nuevos')
```
